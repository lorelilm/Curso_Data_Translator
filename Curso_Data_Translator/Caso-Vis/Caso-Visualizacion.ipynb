```python
import json
with open("data/US_category_id.json", "r") as f:data=json.loads(f.read())
data
```

```python
dataN= pd.json_normalize(data, record_path=["items"])
dataN.head()
```

```python
dataN["category_title"] = dataN["snippet.title"]
```

```python
dataN.head()
```

```python
dataN2 = dataN.rename({'id': 'category_id'}, axis='columns')
```

```python
dataN2.head()
```

```python
dataN2 = pd.DataFrame(dataN2[["category_id", "category_title"]])
```

```python
dataN2.head()
```

```python
dfVideos["category_id"] = dfVideos["category_id"].astype("int")
dataN2["category_id"] = dataN2["category_id"].astype("int")
```

```python
dataParte7 = pd.merge(dfVideos, dataN2, how="inner", left_on= "category_id", right_on="category_id")
dataParte7.head()
```

```python
DF_parte7 = pd.DataFrame(dataParte7[["title","channel_title", "category_title", "views", "likes", "dislikes", "comment_count", "description", "Liked", "tags"]])
```

```python
DF_parte7["tag_conteo"] = DF_parte7["tags"].apply(lambda x:x.count("|")+1 if(x!="[none]") else 0)
```

```python
DF_parte7.head()
```
