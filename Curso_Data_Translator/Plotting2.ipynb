## Plotting 2

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
fig = plt.figure(figsize=(12,4))
ax=fig.add_subplot(111)

ax.set_ylabel("Voltaje")
ax.set_title("Curva de PE")

t=np.arange(0.0,1.0,0.05)
s=np.sin(2*np.pi*t)
line=ax.plot(t,s, color="red", lw=5)
```

```python
fig = plt.figure(figsize=(12,4))
t=np.arange(0.0,1.0,0.05)
s=np.sin(2*np.pi*t)

ax1=fig.add_subplot(3,1,1) #Indice 3,1,1
ax1.set_ylabel("Voltaje")
ax1.set_title("Curva de PE")
line1=ax1.plot(t,s+5.0, color="red", lw=5)
line2=ax1.plot(t,s+3.0, color="blue", lw=3)
line3=ax1.plot(t,s, color="yellow", lw=1)

ax2=fig.add_subplot(3,1,2) #Indice 3,1,2
ax2.set_ylabel("Voltaje")
ax2.set_title("Curva de PE")
line1=ax2.plot(t,s+5.0, color="red", lw=5, linestyle="dashed")
line2=ax2.plot(t,s+3.0, color="blue", lw=3, linestyle="-.")
line3=ax2.plot(t,s, color="yellow", lw=1, linestyle=":")

ax3=fig.add_subplot(3,1,3) #Indice 3,1,3
ax3.set_ylabel("Voltaje")
ax3.set_title("Curva de PE")
line1=ax3.plot(t,s+5.0, color="red", lw=2, marker="+")
line2=ax3.plot(t,s+3.0, color="#000000", lw=3, marker="o")
line3=ax3.plot(t,s, color="orange", lw=1, marker="^")

plt.subplots_adjust(wspace=0.5, hspace=0.3)
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig= plt.figure(figsize=(12,12))
ax=fig.add_subplot(111,projection="3d")

for c,z in zip(["red","green","blue","yellow"],[0,10,20,30]):
    xs= np.arange(20)
    ys= np.random.rand(20)
    
    cs=[c]*len(xs)
    cs[0]="orange" #Color al primero conjunto
    ax.bar(xs,ys,zs=z, zdir="y", color=cs, alpha=0.7)
    
a= ax.set_xlabel("X")
b= ax.set_ylabel("Y")
c= ax.set_zlabel("Z")
```

```python
fig= plt.figure(figsize=(6,6))
with plt.style.context("dark_background"):
    N=4
    x= np.linspace(0,N)
    ncolors=len(plt.rcParams["axes.prop_cycle"])
    shift=np.linspace(0,N, ncolors, endpoint=False)
    for l in shift:
        plt.plot(x,np.sin(x+l), "--")
```

```python
fig= plt.figure(figsize=(6,6))
with plt.style.context("bmh"):
    N=4
    x= np.linspace(0,N)
    ncolors=len(plt.rcParams["axes.prop_cycle"])
    shift=np.linspace(0,N, ncolors, endpoint=False)
    for l in shift:
        plt.plot(x,np.sin(x+l), "--")
```

```python
import seaborn as sns
x=np.arange(0,4,0.1)
y=np.sin(x)

df=pd.DataFrame({"x":x, "sinx":y, "cosx":np.cos(x)})
df=df.set_index("x")

grid=sns.lmplot(x="x", y="sinx", data=df.reset_index())
grid.axes[0,0].hlines(0,-1,5,) #Mover los ejes
ax=grid.axes
```

```python
sns.set(style="dark")

rs= np.random.RandomState(4)
pos= rs.randint(-1,2,(20,5)).cumsum(axis=1)
step= np.tile(range(5),20)
walk= np.repeat(range(20),5)
df= pd.DataFrame(np.c_[pos.flat, step, walk], columns=["pos","step","walk"])

grid=sns.FacetGrid(df,col="walk", hue="walk", col_wrap=5, height=3)

#Linea en Y=0
grid.map(plt.axhline, y=0, ls=":")
#Linea con o
grid.map(plt.plot, "step", "pos", marker="o", ms=4)
```
