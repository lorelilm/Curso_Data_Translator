```python
import numpy as np
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
conn = sqlite3.connect("hr.sqlite"); conn
```

## Parte 1

Mostrar los nombres (first_name, last_name) utilizando el alias “first name”, “last name” de la tabla de empleados

```python
sql = """
select first_name as 'first name', last_name as 'last name'
from employees
"""
pd.read_sql_query(sql, conn)
```

Obtener el “departmet ID” único de la tabla de empleados

```python
df1 = pd.read_sql_query("select department_id from employees", conn); df1.head(2)
```

Obtener todos los detalles de los empleados de la tabla de empleados ordenados por nombre, de manera descendente

```python
sql = """
select * 
from employees
group by first_name
order by first_name desc
"""
pd.read_sql_query(sql, conn)
```

Obtener nombre, apellido, salario y la remuneración de todos los empleados (la remuneración se calcula como el 12% del salario)

```python
sql = """
select first_name as nombre, last_name as apellido, format("$%,d", salary) as salario, format("$%,d",(salary * .12)) as remuneracion 
from employees
"""
pd.read_sql_query(sql, conn)
```

Obtener el  “employee ID”, nombre, apellido y salario ordenados de manera ascendente por el salario.

```python
sql = """
select employee_id as "employee ID", first_name as nombre, last_name as apellido, format("$%,d", salary) as salario 
from employees
order by salary asc
"""
pd.read_sql_query(sql, conn)
```

Obtener el total del salario pagado a los empleados.

```python
df2 = pd.read_sql_query("select format('$%,d', (sum(salary))) as 'Total Salario' from employees", conn); 
df2
```

Obtener el salario máximo y mínimo de la tabla de empleados.

```python
df3 = pd.read_sql_query("select format('$%,d', (max(salary))) as 'Salario Max', format('$%,d', (min(salary))) as 'Salario Min' from employees", conn); 
df3
```

Obtener el salario medio y el número de empleados de la tabla de empleados.

```python
sql = """
select format('$%,d',(avg(salary))) as "Salario medio", count(employee_id) as "Numero de empleados"
from employees
"""
pd.read_sql_query(sql, conn)
```

Obtener el número de puestos de trabajo disponibles en la tabla de empleados.

```python
df4 = pd.read_sql_query("select count(job_id) as 'Puestos Disponibles' from employees", conn); 
df4
```

## Parte 3

Obtener el apellido, el cargo y el salario de todos los empleados cuyo trabajo es el de “Programmer” o “Shipping Clerk”, y cuyo salario no sea igual a 4,500, 10,000 o 15,000 dólares.

```python
sql = """
select em.last_name, jo.job_title, em.salary
from employees em 
inner join jobs jo on jo.job_id=em.job_id
where (jo.job_title='Programmer' or jo.job_title='Shipping Clerk') and em.salary!='4500'
"""
pd.read_sql_query(sql, conn)
```

Obtener los apellidos de los empleados cuyos nombres tienen exactamente 6 caracteres.

```python
sql = """
select last_name
from employees 
where last_name like '______'
"""
pd.read_sql_query(sql, conn)
```

Obtener los apellidos de los empleados que tienen "e" como tercer carácter.

```python
sql = """
select last_name
from employees 
where last_name like '__e%'
"""
pd.read_sql_query(sql, conn)
```

Obtener el salario máximo de un empleado que trabaja como “Programmer”.

```python
sql = """
select em.first_name, em.last_name, max(em.salary) as salary, jo.job_title
from employees em
inner join jobs jo on jo.job_id=em.job_id
where jo.job_title='Programmer'
"""
pd.read_sql_query(sql, conn)
```

Obtener el salario medio y el número de empleados que trabajan en el departamento 90.

```python
sql = """
select avg(em.salary) as salary, count(em.employee_id)
from employees em
inner join departments de on de.department_id=em.department_id
where em.department_id='90'
"""
pd.read_sql_query(sql, conn)
```

Obtener el salario más alto, el más bajo, la suma y la media de todos los salarios.

```python
sql = """
select max(salary) as 'Salario mas alto', min(salary) as 'Salario mas bajo', sum(salary) as 'Suma de salarios', avg(salary) as 'Salario promedio'
from employees 
"""
pd.read_sql_query(sql, conn)
```

Obtener el número de empleados con el mismo puesto.

```python
sql = """
select jo.job_title, count(em.employee_id)
from employees em
inner join jobs jo on jo.job_id=em.job_id
group by jo.job_title
"""
pd.read_sql_query(sql, conn)
```

Obtener la diferencia entre el salario más alto y el más bajo.

```python
df_smax = pd.read_sql_query(" select max(salary) as 'Salario max'from employees", conn);
df_smax
```

```python
df_smin = pd.read_sql_query(" select min(salary) as 'Salario min' from employees", conn);
df_smin
```

```python
df_stotal = df_smax.join(df_smin)
df_stotal
```

```python
df_stotal['Diferencia de salario'] = df_stotal['Salario max'] - df_stotal['Salario min']
df_stotal
```

Obtener el “manager ID” y el salario del empleado con menor sueldo reportando a ese “manager''.

```python
sql = """
select de.manager_id as 'Manager', min(em.salary) as 'Salario minimo'
from employees em
inner join departments de on de.department_id=em.department_id
group by de.manager_id
"""
pd.read_sql_query(sql, conn)
```

## Parte 4

Obtener el “department ID” y el total de salarios a pagar en cada departamento.

```python
sql = """
select de.department_id as 'Departmento', sum(em.salary) as 'Gasto en salarios'
from employees em
inner join departments de on de.department_id=em.department_id
group by de.department_id
"""
pd.read_sql_query(sql, conn)
```

Obtener el salario medio de cada ID de puesto, excluyendo “Programmer”.

```python
sql = """
select jo.job_title as 'Puesto', avg(em.salary) as 'Salario medio'
from employees em
inner join jobs jo on jo.job_id=em.job_id
where not jo.job_title='Programmer'
group by jo.job_title
"""
pd.read_sql_query(sql, conn)
```

Obtener el job ID y el salario máximo de los empleados que su salario máximo es mayor o igual a $4,000.

```python
sql = """
select jo.job_id as 'Job ID', max(em.salary) as 'Salario max'
from employees em
inner join jobs jo on jo.job_id=em.job_id
where em.salary>='4000'
group by jo.job_id
order by max(em.salary) asc
"""
pd.read_sql_query(sql, conn)
```

Obtener el salario medio de todos los departamentos que emplean a más de 10 trabajadores.

```python
df_sdep = pd.read_sql_query("select de.depart_name as 'Department', count(jo.job_id) as 'Trabajadores', avg(em.salary) as 'Salario medio' from employees em inner join jobs jo on jo.job_id=em.job_id inner join departments de on de.department_id=em.department_id group by de.depart_name", conn);
df_sdep
```

```python
df_sdep2 = df_sdep[df_sdep['Trabajadores'] > 10]
df_sdep2
```

Obtener nombre, apellido y  salario de los empleados que tienen un salario más alto que el empleado cuyo apellido es “Bull”.

```python
sql = """
select first_name as 'Nombre', last_name 'Apellido', salary as 'Salario'
from employees em
where salary > (select salary from employees where last_name='Bull')
group by last_name
order by salary asc
"""
pd.read_sql_query(sql, conn)
```

Obtener nombre y apellido de todos los empleados que trabajan en el departamento de IT.

```python
df_sdep = pd.read_sql_query("select em.first_name as 'Nombre', em.last_name as 'Apellido', de.depart_name as 'Department' from employees em inner join departments de on de.department_id=em.department_id where depart_name='IT'", conn);
df_sdep
```

Obtener nombre y apellido de los empleados que tienen un “manager” que trabaja en un departamento con sede en EUA.

```python
sql = """
select first_name as 'Nombre', last_name 'Apellido', co.country_name as 'Sede'
from employees em
inner join departments de on de.department_id=em.department_id
inner join locations lo on lo.location_id=de.location_id
inner join countries co on co.country_id=lo.country_id
where co.country_name='United States of America' and em.manager_id<>'0' and em.employee_id not in (select manager_id from employees)
"""
pd.read_sql_query(sql, conn)
```

Obtener nombre y apellido de los empleados que son “managers”.

```python
sql = """
select first_name as 'Nombre', last_name 'Apellido', job_title
from employees em
inner join jobs jo on jo.job_id=em.job_id
where employee_id in (select manager_id from employees)
"""
pd.read_sql_query(sql, conn)
```

Obtener nombre, apellido y salario de los empleados cuyo salario es mayor que el salario medio.

```python
sql = """
select first_name as 'Nombre', last_name 'Apellido', salary as 'Salario'
from employees em
where salary > (select avg(salary) from employees)
"""
pd.read_sql_query(sql, conn)
```

Obtener nombre, apellido y salario de los empleados que ganan más que el salario medio y que trabajan en cualquier área de IT.

```python
sql = """
select em.first_name as 'Nombre', em.last_name 'Apellido', em.salary as 'Salario', de.depart_name as 'Departamento'
from employees em
inner join departments de on de.department_id=em.department_id
where salary > (select avg(salary) from employees) and depart_name='IT'
"""
pd.read_sql_query(sql, conn)
```

Obtener nombre, apellido y salario de los empleados que ganan más que el Sr. Bell.

```python
sql = """
select first_name as 'Nombre', last_name 'Apellido', salary as 'Salario'
from employees em
where salary > (select salary from employees where last_name='Bell')
group by last_name
order by salary asc
"""
pd.read_sql_query(sql, conn)
```
